```python
import random
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

## Random init the k-armed-bandits

```python
k = 5
maximal_mean = 30
maximal_var = 5

rewards_paramters = {}
for i in range(0,k):
    rewards_paramters[str(i)] = [random.randrange(1,maximal_mean), random.randrange(1,maximal_var)]
```

```python
def i_choose_number(lever) : 
    return(random.gauss(rewards_paramters[str(lever)][0],rewards_paramters[str(lever)][1]))
```

## Viz the effiency by lever 

```python
df = pd.DataFrame(columns=["lever","reward"])

for i in range(0,k*1000) :
    lever = random.randrange(0,k)
    value = i_choose_number(lever)
    df.loc[i] = [lever, value]
```

```python
boxplot = df.boxplot(column=['reward'],by='lever')
```

## Bandit

```python
def bandit(epsilon, Q = [0] * k ,N = [0] * k):
    #print(Q)
    total_reward = [0]
    action = []
    reward = []
    for i in range (0,10000):
        #print("début itération :  " + str(i))
        #print("\t"+ str(Q))
        greedy = random.uniform(0, 1)
        #print("\ton tire aléatoirement {}. 1 - Epsilon vaut {}".format(greedy,1-epsilon))
        if (greedy > epsilon):
        #    print("\ton choisit la solution greddy : ")
            lever_choose = Q.index(max(Q))
        else : 
            #print("\ton tire aléatoirement une solution")
            lever_choose = random.randrange(0,k)
        #print(lever_choose)
        r = i_choose_number(lever_choose)
        N[lever_choose] += 1
        Q[lever_choose] = Q[lever_choose] + (1/N[lever_choose]) * (r - Q[lever_choose])
        total_reward.append(total_reward[-1] + r)
        action.append(lever_choose)

    return([total_reward, action, reward])
```

```python
b1 = bandit(0)
b2 = bandit(0.01)
b3 = bandit(0.1)
b4 = bandit(1)
```

#### final result

```python
print("le premier bandit finit avec un reward de :" + str(b1[0][-1]))
print("le deuxième bandit finit avec un reward de :" + str(b2[0][-1]))
print("le troisème bandit finit avec un reward de :" + str(b3[0][-1]))
```

```python
plt.plot(b1[0], label="Bandit greedy")
plt.plot(b2[0], label="Bandit qui cherche un peu")
plt.plot(b3[0], label="Bandit qui cherche plus")
plt.plot(b4[0], label="Bandit qui ne fait que chercher")
plt.xlabel('nb tirage')
plt.ylabel('Gain')
plt.legend()    
plt.show()
```
